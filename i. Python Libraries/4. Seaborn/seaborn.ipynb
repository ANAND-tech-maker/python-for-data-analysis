```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))  # categorical column - univariate
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
        
```

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['count', 'min', 'max', 'mean', 'median', 'std', 'skew'])) # numerical column - univariate
        print()
```

```python
discrete_df = df.select_dtypes(include=['object']) # select string data types

numerical_df = df.select_dtypes(include=['float64', 'int64']) # select numerical columns 
```

```python
discrete_univariate_analysis(discrete_df) # 4 - string  columns is thire in df 
```

```python
df.dtypes
```

```python
numerical_univariate_analysis(numerical_df) # 3 - numerical columns is thire in df
```
